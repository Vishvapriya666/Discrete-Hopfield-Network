```python
#Aim: Implement a program to store a pattern (1 1 1 -1). Test the network using Discrete Hopfield Net by giving the input with mistakes in First and Second position

#Date of Performance : 12/10/23
#This is the 6th assignment of neural network lab

import numpy as np

# Define the pattern to be stored
pattern = np.array([1, 1, 1, -1])

# Compute the weight matrix using Hebbian learning rule
num_neurons = len(pattern)
weights = np.zeros((num_neurons, num_neurons))

for i in range(num_neurons):
    for j in range(num_neurons):
        if i != j:
            weights[i][j] += pattern[i] * pattern[j]

# Function to update the network state using the given pattern
def update_network_state(input_pattern, weights):
    return np.sign(np.dot(weights, input_pattern))

# Test the network with inputs having mistakes in first and second positions
input_with_mistake_1 = np.array([-1, 1, 1, -1])  # Mistake in the first position
input_with_mistake_2 = np.array([1, -1, 1, -1])  # Mistake in the second position

# Update network state with inputs containing mistakes
output_1 = update_network_state(input_with_mistake_1, weights)
output_2 = update_network_state(input_with_mistake_2, weights)

# Check if the output matches the stored pattern
if np.array_equal(output_1, pattern):
    print("Output with mistake in the first position: Pattern Recognized")
else:
    print("Output with mistake in the first position: Pattern Not Recognized")

if np.array_equal(output_2, pattern):
    print("Output with mistake in the second position: Pattern Recognized")
else:
    print("Output with mistake in the second position: Pattern Not Recognized")
```
